```python
import pandas as pd
```

```python
agg_trans = pd.read_csv('aggregated_transaction.csv')
agg_user = pd.read_csv('aggregated_user.csv')
map_trans = pd.read_csv('map_transaction.csv')
map_user = pd.read_csv('map_user.csv')
top_trans = pd.read_csv('top_transaction.csv')
top_user = pd.read_csv('top_user.csv')
```

```python
top_trans_pers_state_year = top_trans.groupby(['state','year'])['metric.count','metric.amount'].sum()
```

```python
top_user_per_state_year = top_user.groupby(['state','year'])['registeredUsers'].sum()
```

```python
top_trans_pers_state_year = pd.DataFrame(top_trans_pers_state_year)
top_trans_pers_state_year.reset_index(drop = True, inplace=True)
top_trans_pers_state_year
```

```python
top_user_per_state_year = pd.DataFrame(top_user_per_state_year)
top_user_per_state_year.reset_index(inplace=True)
top_user_per_state_year
```

```python
user_vs_txn =  top_trans_pers_state_year.merge(top_user_per_state_year, how='left',left_on=['state','year'],right_on=['state','year'])
```

```python
user_vs_txn.to_excel('user_vs_txn.xlsx')
```

```python
map_user
```

```python
map_user_per_state_year = map_user.groupby(['state','year'])['registeredUsers','appOpens'].sum()
map_user_per_state_year = pd.DataFrame(map_user_per_state_year)
map_user_per_state_year.reset_index(inplace=True)
# map_user_per_state_year
map_txn_per_state_year = map_trans.groupby(['state','year'])['count','amount'].sum()
map_txn_per_state_year = pd.DataFrame(map_txn_per_state_year)
map_txn_per_state_year.reset_index(inplace=True)
# map_txn_per_state_year
state_vs_txn =  map_user_per_state_year.merge(map_txn_per_state_year, how='left',left_on=['state','year'],right_on=['state','year'])
state_vs_txn.to_excel('state_vs_txn.xlsx')
state_vs_txn
```

```python
map_user_per_district_state_year = map_user.groupby(['state','name','year'])['registeredUsers','appOpens'].sum()
map_user_per_district_state_year = pd.DataFrame(map_user_per_district_state_year)
map_user_per_district_state_year.reset_index(inplace=True)
#map_user_per_district_state_year
map_txn_per_district_state_year = map_trans.groupby(['state','name','year'])['count','amount'].sum()
map_txn_per_district_state_year = pd.DataFrame(map_txn_per_district_state_year)
map_txn_per_district_state_year.reset_index(inplace=True)
#map_txn_per_district_state_year
district_vs_txn =  map_user_per_district_state_year.merge(map_txn_per_district_state_year, how='left',left_on=['state','name','year'],right_on=['state','name','year'])
district_vs_txn.to_excel('district_vs_txn.xlsx')
district_vs_txn
```
